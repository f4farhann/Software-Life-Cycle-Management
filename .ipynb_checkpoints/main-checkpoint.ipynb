### Log folder

```python
log_dir = "/home/jovyan/sddc_lcm"
log_files = []
```

```python
time_stamp_pattern = r'\[(\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2} -\d{4})]'
host_pattern = r'(^\S+\.[\S+\.]+\S+)\s'
logger_file_pattern = r'(^\S+\.[\S+\.]+\S+)\s'
has_error_pattern = r''
```

```python
for root, dir_, files in os.walk(log_dir):
    for dir__ in dir_:
        dir__path = os.path.join(root, dir__)
        # print(dir__path)
        # os.system('ls %s' % dir__path)
    for f in files:
        fpath = os.path.join(root, f)
        if ".log" in fpath:
            log_files.append(fpath)
        
```

```python
print(log_files)
```

```python

critical_lines = []
critical_words = ["running", "executing"]
log_file = log_files[0]

with open(log_file) as f:
    f = f.readlines()

for line in f:
    for phrase in critical_words:
        if phrase in line:
            critical_lines.append(line)
            
for line in critical_lines:
    # pass
    print(line)
```
